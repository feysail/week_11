```python
# Plot closing price over time
def plot_closing_price(data,company):
    plt.figure(figsize=(14, 7))
    plt.plot(data, label=company+ " "+ 'Closing Price', color='blue')
    plt.title('TSLA Closing Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_closing_price(data_bnd['Close'],'BND')
```

```python
plot_closing_price(data_tsla['Close'],'TSLA')
```

```python
plot_closing_price(data_spy['Close'],'SPY')
```
